De esta manera vamos a verificar que no hayan registros con datos vacios, si este es el caso, debemos eliminar dichos registros o reemplazarlos. 

```python
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

Aqui confirmamos que no hay un solo registro que no este completo. La variable que estaré tomando en cuenta para este análisis será la columna de 'Output' que muestra el resultado. Pueden ser tres resultados: abandonada, graduado, en curso. Las convertire en numeros para realizar un analisis de correlacion.

```python
estadocivil = {1:'Soltero', 2:'Casado', 3:'Viuda/o', 4:'Divorciado', 5:'Union libre', 6:'Legalmente separados'}
df['Marital status'] = df['Marital status'].replace(estadocivil)
```

```python
counts = df.groupby(['Marital status', 'Output']).size().reset_index(name='Count')
counts['Output'] = counts['Output'].astype(str)
plt.figure(figsize=(7,3))
sns.barplot(data = counts, x='Marital status', y ='Count', hue='Output')
plt.ylabel('Cantidad de registros')
plt.xlabel('Estado civil')
plt.title('Distribucion de registros por estado civil y resultados')
plt.xticks(rotation=30)
plt.show()
```

Las personas solteras son la mayor parte de los registros precedidos por los casados y divorciados.

```python
counts = df.groupby(['Application mode', 'Output']).size().reset_index(name='Count')
counts['Output'] = counts['Output'].astype(str)
plt.figure(figsize=(6,3))
sns.barplot(data = counts, x='Application mode', y ='Count', hue='Output')
plt.ylabel('Cantidad de registros')
plt.xlabel('Modo  de aplicacion')
plt.title('Distribucion de registros por modo  de aplicacion y resultados')
plt.xticks(rotation=30)
plt.show()
```

```python
y = np.asarray(df['Output'])
y[0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
print('Train set:', X_train.shape, y_train.shape)
print('Train set:', X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat,pos_label=0, average='weighted')
```
